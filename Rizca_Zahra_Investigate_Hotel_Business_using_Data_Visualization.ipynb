# Data Preprocessing

#### Handle Null Columns

```python
# first check which columns have null
df.isnull().sum()
```

```python
## There are 4 columns with null values

# Fill children with 0, indicates there's no child when booking the hotel
df['children'] = df['children'].fillna(0)

# Fill city with 'unknown', because we're not sure from where city people booking the hotel
df['city'] = df['city'].fillna('unknown')

#Agent & company column with 0, because both columns are numeric and the minumum agent value is 1 and maksimum company's values is 6
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)

df.isnull().sum()
```

## replace the value that does not match

```python
# Change 'Undefined' value in 'meal' with 'no meal'

df['meal'].replace('Undefined', 'No Meal', inplace=True)
```

## Delete unnecessary data

```python
# make total guest data 
df['guest'] = df['adults'] + df['children'] + df['babies']
print('Data with 0 guest: {0} out of {1} all data'.format(df[df['guest'] == 0].shape[0], df.shape[0]))

# we will deletee data with 0 guest
df_prep = df[df['guest'] > 0].copy()
df_prep.shape
```

```python
print("before pre-processing: ", df.shape[0])
print("after pre-processing: ", df_prep.shape[0])
```

```python
# Copy from pre processing
df1 = df_prep.copy()

# Membuat data agregasi untuk menghitung jumlah pemesanan hotel per bulan berdasarkan tipe hotell
df1_gr = df1.groupby(['hotel', 'arrival_date_month'])['arrival_date_year'].agg(['nunique', 'count']).reset_index()
df1_gr.columns = ['hotel_type', 'arrival_month', 'unique_year', 'num_booking']
df1_gr
```

It can be seen that October and September data get from 3 years, and the other months only from 2 years.

```python
#normalize to get average booking in month in different year
df1_gr['average_num_booking'] = round(df1_gr['num_booking']/df1_gr['unique_year'])
df1_gr.head()
```

```python
#order month
ordered_months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                  'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df1_gr['month'] = df1_gr['arrival_month'].str[:3] # get 3 words
df1_gr['month'] = pd.Categorical(df1_gr['month'], categories=ordered_months, ordered=True)

df1_gr.head()
```

```python
plt.figure(figsize=(15,10)) # mengatur ukuran figure
sns.lineplot(x='month', y='average_num_booking', hue='hotel_type', 
             size="hotel_type", sizes=(2.5, 2.5), data=df1_gr) # plot awal menggunakan lineplot dari library seaborn

plt.tick_params(axis='both', which='major', labelsize=14) # memperbesar ukuran x-y axis label
plt.grid() # menambahkan gridline
plt.legend(title='Hotel Type', title_fontsize=15, prop={'size':13}) # mengatur judul dan ukuran font pada legenda

plt.xlabel('Arrival Month', fontsize=15) # mengatur title pada x-axis 
plt.ylabel('Average Number of Booking', fontsize=15) # mengatur title pada y-axis
plt.ylim(0, 4800) # membatasi y axis

plt.axvline(4, ls='--', color='grey') # membuat garis vertikal untuk menghighlight insight
plt.axvline(6, ls='--', color='grey') # membuat garis vertikal untuk menghighlight insight
plt.stackplot(np.arange(4,7,1), [[4800]], color='grey', alpha=0.3) # memberikan blok warna pada area yang dihighlight 2 garis vertikal

plt.axvline(9, ls='--', color='grey') # membuat garis vertikal untuk menghighlight insight
plt.axvline(11, ls='--', color='grey') # membuat garis vertikal untuk menghighlight insight
plt.stackplot(np.arange(9,12,1), [[4800]], color='grey', alpha=0.3) # memberikan blok warna pada

plt.text(x=-0.5, y=5000, s="Both Hotels Have More Guests During Holiday Season", 
         fontsize=20, fontweight='bold') # memberikan judul yang informatif
#plt.text(x=-0.5, y=4850, s="City Hotel has the decreased guests in August and September, while both hotels have less \ncustomer during not holiday season (Jan-Mar)", 
         #fontsize=18) # memberikan keterangan tambahan atas judul

plt.tight_layout() # mengatur layout dari visualisasi agar tidak terpotong
plt.savefig('/content/gdrive/MyDrive/Mini Project/Investigate Business Hotel using Data Visualization.png', dpi=200); # menyimpan gambar
```

# Impact Analysis of Stay Duration on HotelBookings Canccelation Rates

```python
df2 = df_prep.copy()
df2["total_nights"] = df2["stays_in_weekend_nights"] + df2["stays_in_weekdays_nights"] # mencari total malam yang dipesan

sns.histplot(data=df2, x="total_nights", binwidth=5, color="blue")
```

Distribution of total_nights is right skewed. So it need to be grouped to make the distribution more normal. 

```python
df2['total_nights_group'] = np.where(df2['total_nights'] > 21, 21, 
                                       np.where(df2['total_nights'] > 14, 15, df2['total_nights']))
df2_gr = df2.groupby(['hotel', 'is_canceled', 'total_nights_group'])['arrival_date_month'].count().reset_index(name='num_bookings')
df2_pivot = df2_gr.pivot_table(index=['hotel', 'total_nights_group'], columns='is_canceled', values='num_bookings').reset_index()
df2_pivot['pct_canceled'] = round(df2_pivot[1]*100.0/(df2_pivot[0]+df2_pivot[1]), 2)

df2_pivot.sample(10)
```

```python
plt.figure(figsize=(15,10))
sns.regplot(x='total_nights_group', y='pct_canceled', data=df2_pivot[df2_pivot['hotel'] == 'City Hotel'], scatter=False, label='Trend City Hotel', color='red')
sns.regplot(x='total_nights_group', y='pct_canceled', data=df2_pivot[df2_pivot['hotel'] == 'Resort Hotel'], scatter=False, label='Trend Resort Hotel', color='orange')
sns.barplot(x='total_nights_group', y='pct_canceled', hue='hotel', data=df2_pivot, alpha=0.75, palette = 'BuGn_r')

plt.tick_params(axis='both', which='major', labelsize=14) # memperbesar ukuran x-y axis label, 
plt.grid() # menambahkan gridline
plt.legend(title='Hotel Type', title_fontsize=15, prop={'size':13}) # mengatur judul dan ukuran font pada legenda

plt.xticks(range(0,17,1), labels=[str(x) for x in range(0,15,1)]+["15-21",">21"]) # mengatur customized-label pada x-axis
plt.xlabel('Total Nights Booked', fontsize=15) # mengatur title pada x-axis 
plt.ylabel('Percentage of Booking Canceled (%)', fontsize=15) # mengatur title pada y-axis
plt.ylim(0, 100) # membatasi y axis
plt.xlim(0.5, 16.5) # membatasi x axis

plt.text(x=0.5, y=105, s="Total Nights Have Positive Trend on Percentage of Booking Canceled", 
         fontsize=20, fontweight='bold') # memberikan judul yang informatif

plt.tight_layout() # mengatur layout dari visualisasi agar tidak terpotong
plt.savefig('/content/gdrive/MyDrive/Mini Project/Investigate Business Hotel using Data Visualization.png', dpi=200); # menyimpan gambar
```

```python
df4 = df_prep.copy()

sns.histplot(data=df4, x='lead_time', bins=50, color='red')
```

Terlihat bahwa kolom lead_time skewed right, sehingga akan dilakukan pengelompokan

```python
df4['lead_time'] = np.where(df4['lead_time'] > 400, 401, df4['lead_time']) #lead time dengan value cukup besar(di atas 400) dikategorisasikan menjadi satu grup
bins = list(np.arange(0, max(df4['lead_time']) + 30, 30))
labels = ['{0} -{1}'.format(str(bins[x]), str(bins[x+1])) for x in range(len(bins[:-1]))]
df4['lead_time_group'] = pd.cut(df4['lead_time'], bins=bins, include_lowest=True, labels=labels) #mengelompokan lead_time untuk setiap kelipatan 30
```

```python
#agregasi menghitung jumlah pemesanan dibatalkan berdasarkan group lead_time & tipe hotel
df4_booking = df4.groupby(['hotel', 'lead_time_group', 'is_canceled'])['hotel'].count().reset_index(name='num_bookings')
df4_pivot = df4_booking.pivot_table(index=['hotel', 'lead_time_group'], columns='is_canceled', values='num_bookings').reset_index()
df4_pivot.columns = ['hotel_type', 'lead_time_group', 'not_canceled', 'canceled']
df4_pivot['all_bookings'] = df4_pivot['not_canceled'] + df4_pivot['canceled']
df4_pivot['prcnt_canceled'] = df4_pivot['canceled'] * 100/(df4_pivot['all_bookings'])

df4_pivot.head()
```

```python
plt.figure(figsize=(15,10)) #mengatur ukuran figure
sns.barplot(x='lead_time_group', y='prcnt_canceled', hue='hotel_type', data=df4_pivot, palette = 'BuGn_r') 

plt.tick_params(axis='both', which='major', labelsize=12) #memperbesar ukuran axis x-y
plt.grid() # menambahkan gridline
plt.legend(title='Hotel Type', title_fontsize=14, loc=9)

plt.xlabel('Lead Time Groups (by 30 days)', fontsize=14) #mengatur title pada x-axis
plt.ylabel('Percentage of Booking Canceled(%)', fontsize=14) #mengatur title pada y-axis
plt.ylim(0,100)
plt.xlim(-0.5, 13.5)

#plt.axvline(0.5, ls='--', color='red') #membuat garis vertikal untuk menghighlight insight
plt.text(x=-0.2, y=80, s='1 Month', fontsize=18, rotation=90) #menambahkan teks keterangan
plt.stackplot(np.arange(-0.5, 1.5,1), [[100]], color='grey', alpha=0.3) #memberi blok warna pada area yang di highlight 2 garis vertikal


#plt.axvline(0.5, ls='--', color='red') #membuat garis vertikal untuk menghighlight insight
plt.text(x=10, y=90, s='Around 1 year', fontsize=18) #menambahkan teks keterangan
plt.stackplot(np.arange(8.5, 14.5,1), [[100]], color='grey', alpha=0.3) #memberi blok warna pada area yang di highlight 2 garis vertikal

plt.text(x=-0.5, y=103, s='Within 1 Month of Leads Time, Both Hotels Have the Lowest Canceled Ratio',
         fontsize=20, fontweight='bold', alpha=0.8) #memberikan judul yang informatif

plt.tight_layout()
plt.savefig('/content/gdrive/MyDrive/Mini Project/Investigate Business Hotel using Data Visualization/hotel canceled by leads time.png');
```
